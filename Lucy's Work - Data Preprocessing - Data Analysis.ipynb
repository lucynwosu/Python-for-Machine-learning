#### Calculating and substituting with the Median Value

```python
# Calculate the median value for BloodPressure
BP_Mean = df['BloodPressure'].mean()
# Substitute it in the BloodPressure column of the
# dataset where values are 0
df['BloodPressure'] = df['BloodPressure'].replace(
    to_replace=0, value=BP_Mean)
```

```python
# Calculate the median value for Protien Glucose Concentration
PG_Mean = df['PG_Concentration'].mean()
# Substitute it in the PG Concentration column of the
# dataset where values are 0
df['PG_Concentration'] = df['PG_Concentration'].replace(
    to_replace=0, value=PG_Mean)
```

```python
# Calculate the median value for TSF_Thickness
TSF_Mean = df['TSF_Thickness'].mean()
# Substitute it in the TSF_Thickness column of the
# dataset where values are 0
df['TSF_Thickness'] = df['TSF_Thickness'].replace(
    to_replace=0, value=TSF_Mean)
```

```python
# Calculate the median value for BMI
BMI_Mean = df['BMI'].mean()
# Substitute it in the BMI column of the
# dataset where values are 0 using the replace method
df['BMI'] = df['BMI'].replace(
    to_replace=0, value=BMI_Mean)
```

```python
# Calculate the median value for Insulin
Insulin_Mean = df['SerunInsulin'].mean()
# Substitute it in insulin column of the
# dataset where values are 0
df['SerunInsulin'] = df['SerunInsulin'].replace(
    to_replace=0, value=Insulin_Mean)
```
