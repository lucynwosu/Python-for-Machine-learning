### Machine learning - Principal Component Analysis (PCA)
The main idea of principal component analysis (PCA) is to reduce the dimensionality of a data set consisting of many variables correlated with each other, either heavily or lightly, while retaining the variation present in the dataset, up to the maximum extent. The same is done by transforming the variables to a new set of variables, which are known as the principal components (or simply, the PCs) and are orthogonal, ordered such that the retention of variation present in the original variables decreases as we move down in the order. So, in this way, the 1st principal component retains maximum variation that was present in the original components. The principal components are the eigenvectors of a covariance matrix, and hence they are orthogonal.

#### Explained Variance
The explained variance tells you how much information (variance) can be attributed to each of the principal components. This is important as while 2000 dimensional space can be converted to 2 dimensional space, you lose some of the variance (information) when you do this. By using the attribute explained_variance_ratio_, you can see that the first principal component contains 39.84% of the variance and the second principal component contains 7.74% of the variance. Together, the two components contain 47.59% of the information. This tells us that most of the information is lost therefore we need to find a value for n-component

```python
ns = (pca.explained_variance_ratio_)
```

```python
print(ns)
```

```python
Sum = sum(ns) 
```

```python
print(Sum)
```

#### Choosing the right number of components

```python
pca = PCA().fit(x)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

This curve quantifies how much of the total, 2000-dimensional variance is contained within the first N components. For example, we see that with the digits the first 10 components contain approximately 75% of the variance,30 components contains about 90% of the variance while you need around 60 components to describe close to 100% of the variance.Therefore the number of N-components can be changed accordingly for better result. 

Here see that our two-dimensional projection loses a lot of information (as measured by the explained variance)

### K-means clustering


k-means is  one of  the simplest unsupervised  learning  algorithms  that  solve  the well  known clustering problem. The procedure follows a simple and  easy  way  to classify a given data set  through a certain number of  clusters (assume k clusters) fixed apriori. The  main  idea  is to define k centers, one for each cluster. These centers  should  be placed in a cunning  way  because of  different  location  causes different  result. So, the better  choice  is  to place them  as  much as possible  far away from each other. The  next  step is to take each point belonging  to a  given data set and associate it to the nearest center. When no point  is  pending,  the first step is completed and an early group age  is done. At this point we need to re-calculate k new centroids as barycenter of  the clusters resulting from the previous step. After we have these k new centroids, a new binding has to be done  between  the same data set points  and  the nearest new center. A loop has been generated. As a result of  this loop we  may  notice that the k centers change their location step by step until no more changes  are done or  in  other words centers do not move any more. Finally, this  algorithm  aims at  minimizing  an objective function know as squared error function given by:

                           ‘ci’ is the number of data points in ith cluster. <br>
                           ‘c’ is the number of cluster centers.<br>

* On the left, the C parameter in `scatter()` is the predicted cluster class (`finaldf['klabels'] or kmeans.labels_`). The centroid on the kmeans plot for the selected features (principal component 1 and principal component 2) are shown as well. <br>
* On the right, Label is passed as C parameter to `scatter()`.

```python
f, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2,
                             sharey = True, figsize = (10,6))

# For fitted with kmeans 
ax1.set_title('K Means (K = 2)')
ax1.scatter(x = finalDf['principal component 1'], y = finalDf['principal component 2'], 
            c = finalDf['klabels'], cmap='rainbow')
ax1.scatter(x=centers[:, 0], y=centers[:, 2],
            c='black',s=300, alpha=0.5);

# For original data 
ax2.set_title("Original")
ax2.scatter(x = finalDf['principal component 1'], y = finalDf['principal component 2'], 
            c = finalDf['Label'], cmap='rainbow')
```
