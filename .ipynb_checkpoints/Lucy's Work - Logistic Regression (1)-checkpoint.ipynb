# Data Information

### Dataset

The dataset used is called The Cleveland Heart Disease Dataset provided on the UCI repository scribed below.

#### Attributes 
1.	Age : displays the age of the individual.
2.	Sex : displays the gender of the individual using the following format :
1 = male
0 = female
3.	Chest-pain type : displays the type of chest-pain experienced by the individual using the following format :
1 = typical angina
2 = atypical angina
3 = non  anginal pain
4 = asymptotic
4.	Resting Blood Pressure : displays the resting blood pressure value of an individual in mmHg (unit)
5.	Serum Cholestrol : displays the serum cholestrol in mg/dl (unit)
6.	Fasting Blood Sugar : compares the fasting blood sugar value of an individual with 120mg/dl.
If fasting blood sugar > 120mg/dl then : 1 (true)
else : 0 (false)
7.	Resting ECG : displays resting electrocardiographic results
0 = normal
1 = having ST-T wave abnormality
2 = left ventricular hyperthrophy
8.	Max heart rate achieved : displays the max heart rate achieved by an individual.
9.	Exercise induced angina :
1 = yes
0 = no
10.	ST depression induced by exercise relative to rest : displays the value which is integer or float.
11.	Peak exercise ST segment :
1 = upsloping
2 = flat
3 = downsloping
12.	Number of major vessels (03) colored by flourosopy : displays the value as integer or float.
13.	Thal : displays the thalassemia :
3 = normal
6 = fixed defect
7 = reversible defect

#### Classes
14.	Diagnosis of heart disease : Displays whether the individual is suffering from heart disease or not :
0 = absence
1, 2, 3, 4 = present.

# 2. Data Preprocessing

For easy preprocessing the records containing ? will have to be replaced ny NaN using numpy

```python
#Replacing the ? values
df2 = df.replace("?", np.NaN)
```

```python
df2.tail()
```

Now it can be seen that  ID = 302 feature = Ca has been replaced with NaN. 

```python
df2.isnull().values.any()
```

```python
newMissing_Value = df2.isnull().sum()
newMissing_Value
```

It can be seen that the 'Thal' and 'Ca' columns have 4 and 2 missing values which must be filled before training the model. <br>

### Replacing the missing values using the mean of the feature values in the columns<br>

The mean values for Thal and Ca are found below and then used to **replace the NaNs using the fillna() method**

```python
# Finfing the mean value for Ca
mean_Ca = df2['Ca'].astype(float).mean()
(mean_Ca) 
```

```python
# Finfing the mean value for Thal
mean_Thal = df2['Thal'].astype(float).mean()
(mean_Thal)
```

```python
# The missing values are filled using the mean values

df2.fillna({'Ca':mean_Ca, 'Thal':mean_Thal}, inplace=True)
df2.tail()
```

```python
# Since the mean values are float datatypes, I decided to change Ca and Thal features values all to float so that they can be uniform

convert_dict = {'Ca': float, 
                'Thal': float
               } 
  
df2 = df2.astype(convert_dict) 
print(df2.dtypes) 
```

```python
df2.tail()
```

```python
newMissing_Value2 = df2.isnull().sum()
newMissing_Value2
```

## 4. Building the Prediction Model

This step involves two major activities: Splitting the dataset and training the model

### Split dataset into test and train sets

To achieve this, the **`train_test_split()` method.** is imported from sklearn

```python
from sklearn.model_selection import train_test_split
```

**Separate data into features and target datasets.**

```python
#Drop the 'Num' column which was converted to a Target feature and also the Target column

x = df2.drop('Target', axis = 1)
X = x.drop('Num', axis = 1)
y = df2['Target']
```

```python
y.head()
```

```python
X.head()
```

**Split the data into train-test datasets**

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

#### Fit the training dataset into a logistic regression model

LogisticRegression`is imported and its instance created

```python
from sklearn.linear_model import LogisticRegression
logR = LogisticRegression()
```

```python
logR.fit(X_train,y_train)
```

The X_trian and y_train data is fit into the logistic regression model in order to create a model. 

## 5. Evaluating the Prediction Model

To determine that the prediction model can solve the required problem then we evaluate the model using the test data.

```python
predictions = logR.predict(X_test)
```

```python
df1 = pd.DataFrame({'Actual': y_test, 'Predicted': predictions})
dfPred = df1.head(15)
dfPred
```

#### Classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

#### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test, predictions))
```

#### Accuracy

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predictions))
```

# Result Analysis

**Classifocation Report:**<br>
Precision: This tells how accurate your model is. In other words, you can say, when a model makes a prediction, how often it is correct. The question that this metric answers is, of all patients labeled as having a heart disease how many actually have a heart disease? This Logistic Regression model correctly predicted that patients have a heart disease 86% of the time which is good<br>

Recall:  Recall is the ratio of correctly predicted positive observations to all observations in actual class - Present. The question recall answers is: Of all the patients that truely have a heart disease, how many was labeled correctly? My model has a recall of 0.86 which is good for this model as its above 0.5.<br>

Fi Score: F1 Score is the weighted average of Precision and Recall and this model has a value of 0.86<br>

**confusion matrix**<br>
The confusion matrix is in the form of the array object. The dimension of this matrix is 2*2 because this model is binary classification. I have two classes present and absent. Diagonal values represent accurate predictions, while non-diagonal elements are inaccurate predictions. In the output from using the test data, 43 and 43  are actual predictions, and 6 and 8 are incorrect predictions.


**Accoracy**<br>
 Accuracy is simply a ratio of correctly predicted observation to the total observations. One may think that, One may that high accuracy eans good model. Yes, accuracy is a great measure but with symmetric datasets where values of false positive and false negatives are almost same. TThis model has an accuracy of 86%
