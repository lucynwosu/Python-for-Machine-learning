### Introduction 

Heart disease is estimated to cause about 12 million deaths worldwide every year. The early diagnosis of heart diseases can aid in making decisions on lifestyle changes in high risk patients. This work intends to use the relevant factors(Attibutes) that cause heart disease to predict the overall risk of having a heart disease using decision tree

#### Decision Tree

A decision tree is a decision support tool that uses a tree-like graph or model of decisions and their possible consequences, including chance, event, outcomes, resource costs, and utility. It is one way to display an algorithm that only contains conditional control statements.it is a flowchart-like structure in which each internal node represents a test on an attribute (e.g. whether a coin flip comes up heads or tails), each branch represents the outcome of the test, and each leaf node represents a class label (decision taken after computing all attributes). The paths from root to leaf represent classification rules.



#### Feature Selection

The data features that are used to train a machine learning models have a huge influence on the performance achieved. 
Irrelevant or partially relevant features can negatively impact model performance. Feature selection is a process where you automatically select those features in your data that contribute most to the prediction variable or output in which you are interested.

Having irrelevant features in your data can decrease the accuracy of many models, Three benefits of performing feature selection before modeling your data are:<br>

**Reduces Overfitting: Less redundant data means less opportunity to make decisions based on noise.<br>
Improves Accuracy: Less misleading data means modeling accuracy improves.<br>
Reduces Training Time: Less data means that algorithms train faster.<br>**

This work uses K-best method of feature selection, in which the chi squared (chi^2) statistical test for non-negative features is used to select 3 of the best features from the dataset.

# 4. The Decision Tree Model
## 4a. Building the Decision Tree Prediction Model

This step involves two major activities: Splitting the dataset and training the model

### Split dataset into test and train sets

To achieve this, the **`train_test_split()` method.** is imported from sklearn

```python
from sklearn.model_selection import train_test_split
```

**Separate data into features and target datasets.**

```python
#Drop the 'Num' column which was converted to a Target feature 
x = df2.drop('Target', axis = 1)
X = x.drop('Num', axis = 1)
y = df2['Target']
```

```python
y.head()
```

```python
X.head()
```

**Split the data into train-test datasets**

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

#### Fit the training dataset into a logistic regression model

LogisticRegression`is imported and its instance created

```python
# importing decision tree classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
#Creating instance "dtree" of the classifier 
dtree = DecisionTreeClassifier()
```

```python
#fitting to the training data, the default parameters are fine at the moment!
dtree.fit(X_train,y_train)
```

The X_trian and y_train data is fit into the logistic regression model in order to create a model. 

## 4b. Evaluating the Decision Tree Prediction Model

```python
# doing predictions 
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

#### Classification Report

```python
print(classification_report(y_test,predictions))
```

#### Confusion Matrix

```python
print(confusion_matrix(y_test, predictions))
```

#### Accuracy

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predictions))
```

## 5. Feature Selection + Decision Tree
## 5a. Building the Feature Selection + Decision Tree Prediction Model

```python
from sklearn.feature_selection import SelectKBest, chi2

select_feature = SelectKBest(chi2, k=3).fit(X_train, y_train)
select_feature = SelectKBest(chi2, k=3).fit(X_test, y_test)
```

```python
selected_features_df = pd.DataFrame({'Feature':list(X_train.columns),
                                     'Scores':select_feature.scores_})
selected_features_df.sort_values(by='Scores', ascending=False)
```

```python
X_train_chi = select_feature.transform(X_train)
X_test_chi = select_feature.transform(X_test)
```

```python
X_train_chi[0:3]
```

```python
#Creating instance "dtree" of the classifier 
dtree_Chi_model = DecisionTreeClassifier()
dtree_Chi_model.fit(X_train_chi,y_train)
```

## 5b. Evaluating the Feature Selection + Decision Tree Prediction Model

```python
# doing predictions 
predictions_Chi = dtree_Chi_model.predict(X_test_chi)
```

```python
print(classification_report(y_test,predictions_Chi))
```

```python
print(confusion_matrix(y_test, predictions_Chi))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predictions_Chi))
```

# Result Analysis

**Classifocation Report:**<br>
Precision: This tells how accurate your model is,it tells how often a prediction made by a model is correct. The question that this metric answers is, of all patients labeled as having a heart disease how many actually have a heart disease? The Decision tree model predicted that patients have a heart disease 67% of the time while after feature selection it predits the presence of a heart disease 73% which is an improvement<br>

Recall:  Recall is the ratio of correctly predicted positive observations to all observations in actual class - Present. The question recall answers is: Of all the patients that truely have a heart disease, how many was labeled correctly? My Decision Tree model has a recall of 0.71 which is good as its above 0.5 but the matrix was improved by using feature selection to get a recall of 0.78 <br>

Fi Score: F1 Score is the weighted average of Precision and Recall and this model has a value of 69% and 75%<br>

**confusion matrix**<br>
The confusion matrix is in the form of the array object. The dimension of this matrix is 2*2 because this model is for binary classification. I have two classes present and absent. Diagonal values represent accurate predictions, while non-diagonal elements are inaccurate predictions. In the output from using the test data, The decision tree model has a misclassification rate of 31% will the misclassficication rate reduced to 26%  while using feature selection. 


**Accoracy**<br>
 Accuracy is simply a ratio of correctly predicted observation to the total observations. TThe decision tree model has an accuracy of 69% but with feature selecture the accuracy improved to 75%.
