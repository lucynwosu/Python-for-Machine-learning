#### Database: Heart Disease in Cleveland

This original and full dataset is a part of [UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets.html) and contains 4 databases: ***Cleveland, Hungary, Switzerland, and the VA Long Beach.*** This dataset was donated to UCI Repository in 1988.<br>

Heart Disease in Cleveland: (https://archive.ics.uci.edu/ml/datasets/Heart+Disease). 

Out of the 4 databases, the **Cleveland database** is the only one that has been used by the Machine Learning researchers to this date. In the original database, the "goal" field refers to the presence of heart disease in the patient. It is integer valued from 0 (no presence) to 4. Experiments with the Cleveland database have concentrated on simply attempting to distinguish presence (values 1,2,3,4) from absence (value 0).<br>

I have used an already cleanded the dataset for the missing data. A new column 'target' is also added with N (for 0) and Y (for 1,2,3,4).<br>

The information on the 14 attribute to be used, is provided below: 

* **age**--in years
* **sex**--(1 = male; 0 = female)
* **cp**--chest pain type (1: typical angina, 2: atypical angina, 3: non-anginal, pain 4: asymptomatic)  
* **trestbps** -- resting blood pressure
* **chol**--serum cholesterol in mg/dl
* **fbs**--fasting blood sugar > 120 mg/dl) (1 = true; 0 = false) 
* **restecg**--resting ecg (electrocardiographic) results
* **thalach**--maximum heart rate achieved 
* **exang**--exercise induced angina (1 = yes; 0 = no)
* **oldpeak**--ST depression induced by exercise relative to rest 
* **slope**--the slope of the peak exercise ST segment (1: upsloping, 2: flat, 3: downsloping) 
* **ca**: number of major vessels (0-3) colored by flourosopy 
* **thal**: 3 = normal; 6 = fixed defect; 7 = reversable defect 
* **the predicted attribute** (0, 1, 2, 3 4) -- In the processed dataset, this one is added as a new column 'target' with 'N' for 0 and 'Y' for 1,2,3 & 4.<br>

### Machine Learning Section

### Train Test Split

splitting up the data into a training set and a test set!

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('target',axis=1)
y = df['target']
```

```python
# shift+tab and simply copy
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42)
```

### 1. Logistic Regression

```python
#import LogisticRegression 
from sklearn.linear_model import LogisticRegression
```

```python
#create its instance.
logR = LogisticRegression()
#An Instance of logistic regression, solver='lbfgs' is sued to suppress future warning

logR.fit(X_train,y_train)
```

#### Prediction and Evaluation for Logistic Regression

```python
#make prediction for the X-test set
predictionsL = logR.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
#Evaluate the model using the actual and observed predictions

print(classification_report(y_test,predictionsL))
print(confusion_matrix(y_test,predictionsL))
```

#### Prediction Accuracy for the LoR model

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predictionsL))
```

### 2. Support Vector Machine

```python
#Importing SVC anad create its instance
from sklearn.svm import SVC
svm_model = SVC()

#This takes care of a Futureware message that says gamma is auto-deprecated
svm_model = SVC(gamma='auto') 
```

```python
# training / fitting the model 
svm_model.fit(X_train,y_train)
```

#### Prediction and Evaluation for SVM

```python
predictionsvm = svm_model.predict(X_test)
```

```python
#Evaluate the model using the actual and observed predictions

print(classification_report(y_test,predictionsvm))
print(confusion_matrix(y_test,predictionsvm))
```

#### Prediction Accuracy for the svm model

```python
print(accuracy_score(y_test, predictionsvm))
```

## Conclusion

I have used the following criteria to compare my models but before I actually compare them I will like to explain the terms: 

#### 1. Confusion Matrix 
A performance measurement for machine learning classification problem where output can be two or more classes. It is a table with 4 different combinations of predicted and actual values.



#### 2. True Positive(TP):
Predicted positive and it’s true.
#### 3. True Negative(TN):
Predicted negative and it’s true.
#### 4. False Positive(FP): (Type 1 Error)
Predicted positive and it’s false.
#### 5. False Negative(FN): (Type 2 Error)
Predicted negative and it’s false.

#### 6. precision
Precision is the ability of a classiifer not to label an instance positive that is actually negative. For each class it is defined as as the ratio of true positives to the sum of true and false positives. i.e “for all instances classified positive, what percent was correct?”
#### 7. Recall
Recall is the ability of a classifier to find all positive instances. For each class it is defined as the ratio of true positives to the sum of true positives and false negatives. i.e “for all instances that were actually positive, what percent was classified correctly?”
#### 8. F1 score
The F-score is a weighted harmonic mean of precision and recall such that the best score is 1.0 and the worst is 0.0. Generally speaking, F1 scores are lower than accuracy measures as they embed precision and recall into their computation. The weighted average of F1 should be used to compare classifier models, not global accuracy.
#### 9. Support
Support is the number of actual occurrences of the class in the specified dataset. Imbalanced support in the training data may indicate structural weaknesses in the reported scores of the classifier and could indicate the need for stratified sampling or rebalancing. Support doesn’t change between models but instead diagnoses the evaluation process.
#### 10. Classification accuracy
THis is also known as prediction accuracy is the fraction of predictions a model got right. It is the number of correct predictions made divided by the total number of predictions made, multiplied by 100 to turn it into a percentage.
